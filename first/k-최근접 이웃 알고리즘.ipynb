## K-최근접 이웃

```python
# 도미, 방어 각각 리스트 하나로 합치기
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
# print(fish_data)
```

```python
# fish_data 에서 도미는 35번째 까지고 36 ~ 마지막까지는 방어이므로
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
# 사이킷런 패키지에서 k-최근접 이웃 알고리즘 import
from sklearn.neighbors import KNeighborsClassifier
```

```python
# KNeighborsClassifier 객체 생성
kn = KNeighborsClassifier()
```

```python
# fit(data, target) fit() -> 학습
kn.fit(fish_data, fish_target)
```

```python
# 정확도 확인
kn.score(fish_data, fish_target)
```

```python
# 여기서 x = 30, y = 600 인 생선은?? 도미

# 도미
plt.scatter(bream_length, bream_weight)
# 빙어
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
kn.predict([[30, 600]])
```

```python
# 왜냐하면 가장 가까운 분포된 곳으로 예측하기 때문이다.
# 그래서 x=30, y=600 은 도미로 인식
```

```python
# 데이터 49개로한 모델
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
# 49개중 도미가 35이기때문에 위랑 같은 값나옴
print(35/49)
```
