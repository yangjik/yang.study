**<h2>인공지능 / 머신러닝 / 딥러닝 차이점<br>**
<br>
- 인공지능<br>
사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템.
<br><br>
- 머신러닝<br>
규칙을 일일이 프로그래밍하지 않아도 자동으로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야.<br>
대표 라이브러리는 **사이킷런** 이다.
<br><br>
- 딥러닝<br>
인공 신경망을 기반으로 한 방법들을 통칭함.<br>
대표 라이브러리는 **텐서플로우**, **파이토치** 이다.
<br>


```python
# 도미 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
# 그래프로 데이터 분포도 확인 - 산점도
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('길이')
plt.ylabel('무게')
plt.show()
```

```python
# 빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# 데이터 분포도 확인 - 산점도
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('길이')
plt.ylabel('무게')
plt.show()
```

```python
# 도미, 빙어 데이터 분포도 확인 - 산점도

plt.scatter(bream_length, bream_weight)   # 도미
plt.scatter(smelt_length, smelt_weight)   # 빙어
plt.xlabel('길이')
plt.ylabel('무게')
plt.show()
```

<br>

```python
# 데이터 합치기
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
# 데이터 형태 2차원으로 변경 - zip() 함수 사용
fish_data = [[l, w] for l, w in zip(length, weight)]
fish_data
```

```python
# 리스트에서 1 -> 도미,   0 -> 빙어
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
# K-최근접 이웃 알고리즘 구현한 KNeighborsClassifier import하기
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 클래스 객체 생성
kn = KNeighborsClassifier()
```

```python
# 학습시키기. - kn.fit(x, y)
kn.fit(fish_data, fish_target)
```

```python
# 정확도 확인 - kn.score(x, y)
kn.score(fish_data, fish_target)
```

```python
# 예측 데이터를 전달할때는 앞에 선언한 리스트 차원으로 넣어야된다.
kn.predict([[30, 600]])
```

```python
# 학습시킨x, y축 데이터 확인
print(kn._fit_X)
```

```python
print(kn._y)
```

```python
# 데이터 참고 개수 정하기 default = 5
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
# 전체 데이터 49
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
# 전체 데이터 49개 중 35개가 도미(1)로 지정
print(35/49)
```

결론 : K-최근접 이웃 알고리즘은 기준이 되는 값들에서 가장 가까운 쪽으로 해당 결과를 알려준다.<br>
또는 어떤 데이터에 대한 답을 구할 때 주위의 다른 데이터를 보고 다수를 차지하는 것을 정답으로 사용한다.

```python
# 예제 문제
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
  kn.n_neighbors = KNeighborsClassifier(n_neighbors=n)

  score = kn.score()
```
