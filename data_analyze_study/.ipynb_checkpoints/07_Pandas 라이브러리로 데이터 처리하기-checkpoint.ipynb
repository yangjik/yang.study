#### 위에서 확인한 데이터 처리하기
- 데이터 삭제하기
    - 1. dropna() : 결측치 가진 데이터 삭제 <br><br>
    - 2. subset() : 해당 컬럼을 지정해서 삭제 <br><br>
    - 3. fillna(특정값) : NaN 값을 특정값으로 대체 <br><br>
    - 4. 별도 dict 형태의 데이터를 생성, 없는 데이터를 변경할 컬럼명만 키로 만들고,<br>  변경할 특정 값을 키값으로 넣고, fillna() 함수에 적용해주면 됨.<br><br>
    - 5-1. groupby() : sql 구문의 group by 와 동일, 특정 컬럼을 기준으로 그룹화
    - 5-2. sum() : 그룹으로 되어있는 데이터 합치기<br><br>
    - 6. astype({컬럼명 : 변경할 타입}) : 컬럼 타입변경하기<br><br>
    - 7-1. duplicated() : 중복 행 확인하기 

```python
# 1번
csv = pd.read_csv(file_path + '04-01-2020.csv', encoding='utf-8')
drop_csv = csv.dropna()
drop_csv.head(3)
```

```python
# 2번
csv = pd.read_csv(file_path + '04-01-2020.csv', encoding='utf-8')
subset_csv = csv.dropna(subset=['Confirmed'])
subset_csv.head(2)
```

```python
# 3번
csv = pd.read_csv(file_path + '04-01-2020.csv', encoding='utf-8')
fill_csv = csv.fillna(0)
fill_csv.head(3)
```

```python
# 4번
csv = pd.read_csv(file_path + '04-01-2020.csv', encoding='utf-8')
dict_data = {'Deaths' : 0, 'Recovered' : 0}
dict_csv = csv.fillna(dict_data)
dict_csv.head()
```

```python
# 5
csv = pd.read_csv(file_path + '04-01-2020.csv', encoding='utf-8')
group_csv = csv.groupby('Country_Region').sum()
group_csv.head()
```

중복이 아닐땐 False / 중복일 때 True

drop_ducplicates() : 중복 행 삭제중복값
- 특정 컬럼을 기준으로 중복 행 제거하기
    - subset='특정컬럼'
- 중복된 경우, 처음과 마지막 행 중 어느 행을 남길 것인지 결정하기
    - 처음 : keep='first'(기본값)
    - 마지막 : keep='last'

```python
csv.drop_duplicates(subset='Country_Region', keep='last')
```
