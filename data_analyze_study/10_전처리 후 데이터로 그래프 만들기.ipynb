### 국가명과 iso2 매칭 테이블 읽기(country_info)
- iso2 컬럼값으로 https://www.countryflags.com/ 에서 제공하는 국기 이미지 링크를 얻을 수 있다.
- keep_default_na = False : 디폴트 결측치 변환 데이터를 사용하지 않고, na_values로 지정한 데이터만 결측치로 변환
- na_values='' : 결측치로 변환할 값을 지정


````python
country_info = pd.read_csv('ex_datafile/COVID-19-master/csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv', encoding='utf-8',
                           keep_default_na = False, na_values = '')
country_info
````

````python
country_info[country_info['Country_Region'] == 'Namibia' ]
````

````python
country_info = country_info[['iso2', 'Country_Region']]
country_info
````

````python
country_info.shape
````

````python
country_info = country_info.drop_duplicates(subset='Country_Region', keep='last')
country_info.shape
````

````python
country_info
````

#### 날짜, 국가별 확진자수와 국가별 iso2 값 병합

````python
final_data = pd.merge(df_confirmed, country_info, how = 'left', on = 'Country_Region')
final_data
````

````python
final_data.shape
````

#### 참고: 데이터프레임에 신규 컬럼 추가하기
- 데이터 프레임에 신규 컬럼 추가하기
  - 데이터프레임[신규컬럼] = pd.Series(data=데이터)
    - 데이터는 행에 맞게, [0, 1, 2] 와 같은 형태로 넣을 수 있음
- 기존 데이터프레임 컬럼 값을 기반으로 신규 컬럼 추가하기
  - apply() 함수를 사용
  - 다음 create_flag_link 함수 적용 전에, 아래의 함수로 어떤 값이 인자로 들어오는지 확인해보기 

    ```
    def create_flag_link(row):
        flag_link = 'https://flagcdn.com/48x36/' + row + '.png'
        return flag_link
        
    doc_final_country['Country_Flag'] = doc_final_country['iso2'].apply(create_flag_link)
    ```

````python
def link(row):
    flag_link = 'https://flagcdn.com/48x36/' + row.lower() + '.png'
    return flag_link

final_data['iso2'] = final_data['iso2'].apply(link)
````
