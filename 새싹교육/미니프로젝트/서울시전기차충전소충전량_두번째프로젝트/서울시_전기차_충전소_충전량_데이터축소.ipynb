```python
# 패키지 설정
import pandas as pd
import numpy as np
```

```python
file_path = '/content/drive/MyDrive/새싹교육/미니프로젝트/서울시 전기차 충전소 충전량/한국전력공사_서울시 전기차 충전소 충전량_20220331.xlsx'

df = pd.read_excel(file_path)
```

```python
df['시작_연도'] = df['충전시작시각'].dt.year
df['시작_월'] = df['충전시작시각'].dt.month
df['시작_일'] = df['충전시작시각'].dt.day
df['시작_시'] = df['충전시작시각'].dt.hour
df['시작_분'] = df['충전시작시각'].dt.minute

df['종료_연도'] = df['충전종료시각'].dt.year
df['종료_월'] = df['충전종료시각'].dt.month
df['종료_일'] = df['충전종료시각'].dt.day
df['종료_시'] = df['충전종료시각'].dt.hour
df['종료_분'] = df['충전종료시각'].dt.minute
```

데이터가 너무 많으므로 21년도로 범위 축소.

```python
new_csv = df[df['종료_연도'] < 2022]
```

```python
new_csv.to_csv('/content/drive/MyDrive/새싹교육/미니프로젝트/서울시 전기차 충전소 충전량/21년도_서울시_전기차_충전소.csv')
```
