### 데이터 불러오기 및 데이터 구조 파악

```python
# 패키지 설정
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
import chart_studio.plotly as py
import cufflinks as cf

# cufflinks 서비스 이용하려면 오프라인으로
cf.go_offline(connected = True)
```

### 모델 학습 - 분류 모델 사용
- XGBoost
- Logistic Regression

```python
# 데이터 분리
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    bc_df.drop(columns=['result']), bc_df['result'], test_size=0.2, stratify=bc_df['result'], random_state=42
)
```

```python
train_input.shape, train_target.shape, test_input.shape, test_target.shape
```

#### 로지스틱회귀 - 분류

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

par = {
    'penalty' : ['l2','l1'],
    'C':[0.01, 0.1, 1, 1, 5, 10]
}

lr = LogisticRegression(random_state=42, n_jobs=-1, max_iter=1000, solver='liblinear')
gs = GridSearchCV(lr, cv=3, n_jobs=-1, param_grid=par)

lr_gs = gs.fit(train_input, train_target)
hyper = lr_gs.best_params_
```

```python
# 그리드 서치 결과
pd.DataFrame(lr_gs.cv_results_)
```

```python
hyper['C']
```

```python
lr = LogisticRegression(C=hyper['C'], penalty=hyper['penalty'], random_state=42, n_jobs=-1, max_iter=1000, solver='liblinear')
lr.fit(train_input, train_target)
```

```python
lr_predict = lr.predict(test_input)
```

```python
print('훈련 점수 : ', lr.score(train_input, train_target), '\n테스트 점수 : ', lr.score(test_input, test_target))
```

```python
# 평가 지표 확인
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

#def accuracy(test_target, predict)
con_m = confusion_matrix(test_target, lr_predict) # 
ac_s = accuracy_score(test_target, lr_predict)
pre_s = precision_score(test_target, lr_predict)
re_s = recall_score(test_target, lr_predict)
f1_s = f1_score(test_target,lr_predict)
```

```python
# confustion_matrix - 오차행렬 시각화
sns.heatmap(con_m, annot=True)

plt.title('오차행렬 시각화')
plt.xlabel('참 / 거짓')
plt.ylabel('거짓 - 참')
plt.show()
```

```python
print('정확도 : ', ac_s, '\n정밀도 : ', pre_s, '\n재현율 : ', re_s, '\nF1 : ', f1_s)
```

#### XGBoost - XGBClassifier(분류 모델)

```python
import xgboost as xgb
from xgboost import XGBClassifier
par = {
    'n_estimators' : [1, 10, 100, 200, 400, 600],
    'max_depth' : [1, 2, 3, 4, 5]
}
xgb = XGBClassifier()
gs = GridSearchCV(xgb, param_grid=par, cv=3, n_jobs=-1)
xgb_gs = gs.fit(train_input, train_target)

xgb_cl = XGBClassifier(n_estimators=xgb_gs.best_params_['n_estimators'], max_depth=xgb_gs.best_params_['max_depth'], cv=3, n_jobs=-1)
xgb_cl.fit(train_input, train_target)
xgb_cl_pre = xgb_cl.predict(test_input)
```

```python
# 그리드 서치 결과
pd.DataFrame(xgb_gs.cv_results_)
```

```python
# 평가 지표 확인
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

con_m = confusion_matrix(test_target, xgb_cl_pre) # 
ac_s = accuracy_score(test_target, xgb_cl_pre)
pre_s = precision_score(test_target, xgb_cl_pre)
re_s = recall_score(test_target, xgb_cl_pre)
f1_s = f1_score(test_target,xgb_cl_pre)
```

```python
# confustion_matrix - 오차행렬 시각화
sns.heatmap(con_m, annot=True)

plt.title('오차행렬 시각화')
plt.xlabel('참 / 거짓')
plt.ylabel('거짓 - 참')
plt.show()
```

```python
print('훈련 점수 : ', xgb_cl.score(train_input, train_target), '\n테스트 점수 : ', xgb_cl.score(test_input, test_target))
```

```python
print('정확도 : ', ac_s, '\n정밀도 : ', pre_s, '\n재현율 : ', re_s, '\nF1 : ', f1_s)
```

#### XGBoost - XGBClassifier(분류 모델)

```python
import xgboost as xgb
from xgboost import XGBRegressor

par = {
    'max_depth' : [1, 3, 5, 7, 9],
    'gamma' : [0.01, 0.1, 1, 10, 100],
}

xgb_r = XGBRegressor(random_state=42)
gs = GridSearchCV(xgb_r, param_grid=par, cv=3, n_jobs=-1)
gs.fit(train_input, train_target)
gs.best_params_

xgb_re = XGBRegressor(gamma=gs.best_params_['gamma'], max_depth=gs.best_params_['max_depth'], random_state=42, n_jobs=-1)
xgb_re.fit(train_input, train_target)
xgb_re_pre = xgb_re.predict(test_input)
```

```python
# 그리드 서치 결과
pd.DataFrame(gs.cv_results_)
```

```python
# 평가 지표 확인
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(test_target, xgb_re_pre)
mse = mean_squared_error(test_target, xgb_re_pre)
rmse = mean_squared_error(test_target, xgb_re_pre, squared=False)
r2 = r2_score(test_target, xgb_re_pre)
```

```python
print('훈련 점수 : ', xgb_re.score(train_input, train_target), '\n테스트 점수 : ', xgb_re.score(test_input, test_target))
```

```python
print('MAE : ', mae, '\nMSE : ', mse, '\nRMSE : ', rmse, '\nR2 : ', r2)
```
